```python
# importing relevent library
import pandas as pd
import numpy as np
# import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # all plot will cover on seaborn style
```

```python
dataset = pd.read_csv('../data/cars_revised_data.csv')
dataset.head()
```

```python
# plotting all the three numerical columns
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
sns.scatterplot(x=dataset['year'], y=dataset['selling_price']);
plt.title('Year Vs Price');

plt.subplot(1,2,2)
sns.scatterplot(x=dataset['mileage'], y=dataset['selling_price']);
plt.title('Mileage Vs Price')
```

```python
sns.histplot(dataset['selling_price'], kde=True)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Count')
```

```python
sns.boxplot(x='fuel', y='mileage', data=dataset)
plt.title('Mileage by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage')
```

```python
numeric_cols = ['year', 'selling_price', 'km_driven', 'mileage', 'engine', 'max_power', 'seats']
correlation = dataset[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
```
